# Illustration of H-band magnitude predictor based on PS1 + Gaia, trained using H-band information from VIKING and VVV

### Rationale
For the ELT to operate we will need reference star catalogues with information in the optical and Near-IR. Since not the entire sky that will be accessible by the ELT is covered with deep (few mag beyond 2MASS) NIR imaging surveys, we explore the possibility to predict the missing information using data that is available in the optical. 

When observing a field with respect to the position of a reference star, we note that proper motions of stars are typically of order 10 mas/year. This implies a yearly shift that is substantial compared to the ELT instrumental pixel scales and the width of the diffraction-limited PSF. It is therefore important to consider the proper motions of the reference stars so that different data sets, particularly those taken in different periods, are combined. Gaia overlap is thus essential to be able to track stellar proper motions at the required precision. 

To train a model that predicts H-band fluxes of stars, we use as input magnitudes from Gaia and Pan-STARRS1 (PS1). The model should perform well in regions of different interstellar extinction, so we will train and test it using the VIKING and VVV regions. Those surveys provide the H-band magnitudes that we aim to predict.

To be more precise, we input colours rather than magnitudes, to have a more robust prediction. The final value is a G-H colour, where capital G is the Gaia very broad G-band. The predicted H-band is simply the total G-band mangnitude minus G-H colour.

### Obtaining and Cleaning the data
We run SQL functions to query the CDS/Vizier databases in the VVV and VIKING fields. The data set contains approximately the same number of VIKING and VVV sources. We query the Gaia+PS1 survey in the same fields and match the sources by coordinate.
We require all colours to be measured (no NaNs, which implies detections in all optical bands). A similar cleaning will need to be done on the final data set when this model will be applied.
We also remove stars for which the H-magnitude is brighter than 13.0. That is the approximate saturation limit in the VISTA+VVV surveys. 
Since the connection is not always stable, the scripts automatically restart the connection whenever it is lost (up to 5 times per query).

### Model used
We use a Neural Network to map the input vector (i.e. available colours in the optical), to a single float value that represents the G-H colour. 

To optimise the performance we searched over various NN architectures, either one, two or three fully connected hidden layers. Two layers with 32 units gave a good performance on the cross validation set (0.07 mean absolute error). More complexity (layers and/or neurons) only improved the performance on the training set, and not on the CV set; this is an indication of overfitting. The performance degrades for two layers of 16 units each. Each unit in the hidden layer has a ReLU activation function, with a final single unit with a linear activation to give the G-H colour prediction.

### Performance
Performance is generally good: 0.07 mag mean absolute error. We reach a slightly better performance for VIKING than VVV (highly extincted region of the Galactic centre). 
Even colour outliers (in PS1 or Gaia, as identified using a OneClassSVM classifier) are generally giving a good G-H colour estimate. However, the performance is slighly better for "standard" colours.

# Load function and Modules

```python
# BASIC FUNCTIONS AND MODULES
import seaborn as sns
import sys
import math
import pandas as pd
pd.set_option('max_columns', None)
import numpy as np
import six
from pyvo.dal import tap
import matplotlib
from matplotlib.pylab import plt
import time
import pydot

# Astropy
import astropy
from astropy import units as u
from astropy.coordinates import SkyCoord
from astropy.units import Quantity
from astropy.coordinates.tests.utils import randomly_sample_sphere
from astroquery.esasky import ESASky
from astropy.coordinates import match_coordinates_sky
from astropy.stats import sigma_clipped_stats

#Tensorflow specific
# Make numpy printouts easier to read.
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.layers.experimental import preprocessing


#SQL setup
VIZIER_TAP_CAT = "http://tapvizier.u-strasbg.fr/TAPVizieR/tap"
tapcatvizier = tap.TAPService(VIZIER_TAP_CAT)
def query_Gaia_PS1(ra_search, dec_search, halfboxsize):
    minra = ra_search-halfboxsize
    maxra = ra_search+halfboxsize
    mindec = dec_search-halfboxsize
    maxdec = dec_search+halfboxsize
    query="""SELECT RA_ICRS, DE_ICRS, gaia.Gmag, "BP-RP" as BPmRP, "BP-G" as BPmG, RUWE, epsi, sepsi, 
    ps1.gmag-rmag as gmr, rmag-imag as rmi, imag-zmag as imz, zmag-ymag as zmy, gaia.Gmag-rmag as Gmr,
    "b_rmag" as rmagmin, "B_rmag" as rmagmax, ps1.e_gmag, ps1.gmagStd, ps1.e_rmag, ps1.rmagStd, ps1.e_imag, ps1.imagStd
    FROM "I/350/gaiaedr3" AS Gaia, "II/349/ps1" as PS1
    WHERE RA_ICRS > %f AND RA_ICRS < %f AND DE_ICRS > %f AND DE_ICRS < %f AND objID = PS1"""  % (minra, maxra, mindec, maxdec) 
    res = runqueryvizier(query)
    df_vizier = res.to_table().to_pandas()
    return df_vizier

def query_virac(ra_search, dec_search, halfboxsize):
    minra = ra_search-halfboxsize
    maxra = ra_search+halfboxsize
    mindec = dec_search-halfboxsize
    maxdec = dec_search+halfboxsize
    query="""SELECT RA_ICRS as RAJ2000, DE_ICRS as DEJ2000, Ymag, Jmag, Hmag, Ksmag 
    FROM "II/364/virac" 
    WHERE RA_ICRS > %f AND RA_ICRS < %f AND DE_ICRS > %f AND DE_ICRS < %f"""  % (minra, maxra, mindec, maxdec) 
    res = runqueryvizier(query)
    df_vizier = res.to_table().to_pandas()
    return df_vizier

def query_viking(ra_search, dec_search, halfboxsize):
    minra = ra_search-halfboxsize
    maxra = ra_search+halfboxsize
    mindec = dec_search-halfboxsize
    maxdec = dec_search+halfboxsize
    query="""SELECT RAJ2000, DEJ2000, Yap3 as Ymag, Jap3 as Jmag, Hap3 as Hmag, Ksap3 as Ksmag
    FROM "II/343/viking2" 
    WHERE RAJ2000 > %f AND RAJ2000 < %f AND DEJ2000 > %f AND DEJ2000 < %f"""  % (minra, maxra, mindec, maxdec) 
    res = runqueryvizier(query) 
    df_vizier = res.to_table().to_pandas()
    return df_vizier

def runqueryvizier(query):
    """Make sure the query runs, even if connection is lost. Retry a few times"""
    retry_flag = True
    retry_count = 0
    while retry_flag and retry_count < 5:
        try:
            res = tapcatvizier.search(query=query)
            retry_flag = False
        except Exception as e:
            print (e)
            print ("Retry after 5 sec")
            retry_count = retry_count + 1
            time.sleep(5)
    return res
```

# Fill dataframe for training and testing

```python
# First area of high extinction from VVV (VIRAC)
# We deliberately spread the chosen region over different survey 
# tiles to have a more representative selection of stars
df_master1 = pd.DataFrame({})

boxsize = 0.05
for racent in np.arange(264.,278.,13.9/3.):
    for deccent in np.arange(-29.,-24.,2.4):
        df_VISTA = query_virac(racent,deccent,boxsize).dropna().reset_index().drop(columns='index')
        df_GaiaPS1 = query_Gaia_PS1(racent,deccent,boxsize).dropna().reset_index().drop(columns='index')       
        if len(df_GaiaPS1) > 0.5 and len(df_VISTA) > 0.5:
            c = SkyCoord(ra=df_VISTA['RAJ2000']*u.degree, dec=df_VISTA['DEJ2000']*u.degree)
            catalog = SkyCoord(ra=df_GaiaPS1['RA_ICRS']*u.degree, dec=df_GaiaPS1['DE_ICRS']*u.degree)
            idx, d2d, d3d = catalog.match_to_catalog_sky(c)
            selarr=((d2d.arcsec <0.5))
            df_GaiaPS1.loc[selarr,'matchdistarcsecVISTA']=d2d[selarr].arcsec
            df_GaiaPS1_cut = df_GaiaPS1.loc[selarr.nonzero()].reset_index()
            df_VISTA_cut = df_VISTA.loc[idx[selarr.nonzero()[0]]].reset_index()
            df_new = df_GaiaPS1_cut.merge(df_VISTA_cut, left_index=True, right_index=True, how='right').drop(columns=['index_x','index_y'])
            df_master1 = df_master1.append(df_new)
    
```

```python
# Now area of low extinction from VIKING
# We deliberately spread the chosen region over different survey 
# tiles to have a more representative selection of stars
df_master2 = pd.DataFrame({})

boxsize = 0.25
for racent in np.arange(129.,140.,10.9/6):
    for deccent in np.arange(-2.,2.,3.9/3):
        df_VISTA = query_viking(racent,deccent,boxsize).dropna().reset_index().drop(columns='index')
        df_GaiaPS1 = query_Gaia_PS1(racent,deccent,boxsize).dropna().reset_index().drop(columns='index')       
        if len(df_GaiaPS1) > 0.5 and len(df_VISTA) > 0.5:
            c = SkyCoord(ra=df_VISTA['RAJ2000']*u.degree, dec=df_VISTA['DEJ2000']*u.degree)
            catalog = SkyCoord(ra=df_GaiaPS1['RA_ICRS']*u.degree, dec=df_GaiaPS1['DE_ICRS']*u.degree)
            idx, d2d, d3d = catalog.match_to_catalog_sky(c)
            selarr=((d2d.arcsec <0.50))
            df_GaiaPS1.loc[selarr,'matchdistarcsecVISTA']=d2d[selarr].arcsec
            df_GaiaPS1_cut = df_GaiaPS1.loc[selarr.nonzero()].reset_index()
            df_VISTA_cut = df_VISTA.loc[idx[selarr.nonzero()[0]]].reset_index()
            df_new = df_GaiaPS1_cut.merge(df_VISTA_cut, left_index=True, right_index=True, how='right').drop(columns=['index_x','index_y'])
            df_master2 = df_master2.append(df_new)
    
```

```python
# Combine both in one DataFrame
df_master = pd.DataFrame({})
df_master = df_master.append(df_master1)
df_master = df_master.append(df_master2)

len(df_master)
```

# Clean dataframe and prepare for training

```python
df_master = df_master.reset_index().drop(columns='index')
df_master['GmH'] = df_master['Gmag']-df_master['Hmag']
#brighter than 13 is saturated. Doesn't work for training purposes
before = len(df_master)
df_master = df_master[(df_master['Hmag'] > 13.0) & (df_master['Gmag'] > 13.0)].reset_index().drop(columns='index')
after = len(df_master)
print(f"{100*(before-after)/before:.1f}% of saturated sources removed")

# Some PS1 source indicate significant flux variations compared to the uncertainty expected for their mean magnitude.
# We remove these variable sources before fitting the model.
before = len(df_master)
SEL = ((df_master['gmagStd'] > 0.5) & (df_master['gmagStd'] > 3.*df_master['e_gmag'])) | \
    ((df_master['rmagStd'] > 0.5) & (df_master['rmagStd'] > 3.*df_master['e_rmag'])) | \
    ((df_master['imagStd'] > 0.5) & (df_master['imagStd'] > 3.*df_master['e_imag']))
df_master = df_master[~SEL].reset_index().drop(columns='index')
after = len(df_master)
print(f"{100*(before-after)/before:.1f}% of PS1 problematic sources removed")

# By running the model initially, we find that Viking has a MAGZP problem in a particular sky region. 
# Comment the next two lines to explore this further
SEL = np.sqrt((np.abs(df_master["RA_ICRS"] -130.8)**2) + (np.abs(df_master["DE_ICRS"] -0.6)**2)) > 0.5
df_master = df_master[SEL].reset_index().drop(columns='index')
```

# Prepare data set by splitting between train and test set  

```python
#Random split. Important since VVV and VISTA sources are not mixed
train_dataset = df_master.sample(frac=0.8, random_state=0)
test_dataset = df_master.drop(train_dataset.index)
```

```python
train_dataset.describe().transpose()
```

```python
# Only select relevant features
train_features = train_dataset[['BPmRP','BPmG','gmr','rmi','imz','zmy','Gmr']]
test_features = test_dataset[['BPmRP','BPmG','gmr','rmi','imz','zmy','Gmr']]

train_labels = train_dataset['GmH']
test_labels = test_dataset['GmH']
```

# Set up and train model

We are using Tensorflow to train a Neural Network for a regression task. For a general introduction/tutorial on this use of Tensorflow, see: https://www.tensorflow.org/tutorials/keras/regression
        

```python
# Normalise data. 
# Normalizing the data generally speeds up learning and leads to faster convergence.
normalizer = preprocessing.Normalization(axis=-1)
normalizer.adapt(np.array(train_features))
print(normalizer.mean.numpy())
```

```python
def build_and_compile_model(norm):
    """This sets up the NN. Can try different architectures."""
    model = keras.Sequential([
      norm,
      layers.Dense(32, activation='relu'),
      layers.Dense(32, activation='relu'),
      layers.Dense(1)
    ])

    model.compile(loss='mean_absolute_error',
                optimizer=tf.keras.optimizers.Adam(0.0003))
    return model

def plot_loss(history):
    """Check performance"""
    plt.plot(history.history['loss'], label='loss')
    plt.plot(history.history['val_loss'], label='val_loss')
    plt.ylim([0.05, 0.3])
    plt.xlabel('Epoch /Training step')
    plt.ylabel('Error [Hmag]')
    plt.yscale('log')
    plt.legend()
    plt.grid(True)
```

```python
dnn_model = build_and_compile_model(normalizer)
dnn_model.summary()
```

```python
%%time
history = dnn_model.fit(
    train_features, train_labels,
    validation_split=0.2,
    verbose=0, epochs=400)
```

```python
#Check performance
#32, 32, learning rate 0.0003, i.e. basic model
plot_loss(history)
print(history.history['val_loss'][-1])
```

# Outlier detection with OneClassSVM
The above model shows a reasonable performance but there are some outliers. The following cells check whether these outliers are associated with non-standard optical colours. <br> 
We run an unsupervised outlier detection algorith, OneClassSVM (https://scikit-learn.org/stable/modules/generated/sklearn.svm.OneClassSVM.html), which we let identify the 1% most deviant colours in multi-dimensional (optical) colour space. Then we check whether the H-band magnitude predictions of those explain the outliers.


```python
from sklearn import svm
```

```python
outliers_fraction = 0.01 #The 1% most deviant sources
algorithm = svm.OneClassSVM(nu=outliers_fraction, kernel="rbf", gamma=0.1)
```

```python
algorithm.fit(train_features)
```

```python
outliers = algorithm.predict(train_features)
# Temporary dataframe to keep train_features intact
df_explore_outliers = train_features.reset_index().drop(columns='index')
df_explore_outliers['mark'] = 'normal'
df_explore_outliers.loc[outliers < 0,'mark'] = 'outlier'
print(f"{100*np.sum(outliers < 0)/len(outliers):.1f}% outliers marked")
```

```python
#Show where outliers are in multi-dimensional colour space
sns.pairplot(df_explore_outliers, hue='mark', vars=['Gmr','BPmRP','BPmG'], plot_kws = {'alpha': 0.1}, size=4, diag_kind='kde')
plt.show()
sns.pairplot(df_explore_outliers, hue='mark', vars=['gmr', 'rmi', 'imz', 'zmy'], plot_kws = {'alpha': 0.1}, size=4, diag_kind='kde')
plt.show()
```

```python
#predict outliers on test set based on OneClassSVM model trained on training data, using same features
outliers = algorithm.predict(test_dataset[['BPmRP','BPmG','gmr','rmi','imz','zmy','Gmr']])
print(f"{100*np.sum(outliers < 0)/len(outliers):.1f}% outliers marked")
```

```python
SEL = outliers < 0
plt.scatter(test_dataset[~SEL]['Hmag'],test_dataset[~SEL]['Hmag']-test_dataset[~SEL]['Hmag_predict'],marker='.', label='normal', alpha=0.1)
plt.scatter(test_dataset[SEL]['Hmag'],test_dataset[SEL]['Hmag']-test_dataset[SEL]['Hmag_predict'],marker='.', label='outliers')
plt.xlabel('Hmag_true')
plt.ylabel('Hmag_true-Hmag_predict')
plt.legend()
plt.show()
```

```python
# When flagging optical-colour outliers, a small fraction of G-H outliers can be flagged
bins=np.linspace(-0.5,0.5,100)
plt.hist(test_dataset['Hmag']-test_dataset['Hmag_predict'],density=False,bins=bins, label='normal')
plt.hist(test_dataset[SEL]['Hmag']-test_dataset[SEL]['Hmag_predict'],density=False,bins=bins,alpha=0.5, label='outlier')
plt.xlabel('Hmag_true-Hmag_predict')
plt.ylabel('#')
plt.ylim(0,10)
plt.legend()
plt.show()
```
