# Flag contaminating sources based on Gaia EDR3 parameters

### Rationale

Workbook 2 predicts many stars, that the ELT could use for wavefront sensing, to be binary star systems with angular separation and brightness ratios that may be problematic for ELT operations. On top of this, we expect extended sources (galaxies) to be part of the Gaia catalogue.

On the HST imaging data we have two approaches to flag problematic sources. First, using the CNN that was trained in Workbook 3, we are sensitive to binary stars with small angular separations. It is not specifically trained on extended sources though. Secondly, therefore, we are using a curve-of-growth analysis, i.e. comparing the aperture fluxes in different circular apertures, we can identify extended objects such as galaxies. 

Based on an identification of these problematic sources in HST imaging data, this Notebook explores how these sources can be found by using Gaia data only. This approach would allow us to flag problematic sources whenever HST imaging data is not available (i.e. most of the sky).

### Obtaining and Cleaning the data
For this proof-of-concept we use the same CANDELS data set as we used in the previous Workbook. We only make use of the "real" data here, not the binary stars we created to train the CNN.

### Model used
We use a Convolutional Neural Network (CNN) that was trained in Workbook 3. The "curve-of-growth" analysis is based on aperture fluxes in circular apertures of 5, 10, 20 and 60 pixels from the determined centroid.

### Performance and next steps
We test how well several Gaia EDR3 parameters perform in cleaning the data set. The contamination, estimated from HST, goes down from ~11% to a few percent when using some of these parameters.

A promising next step would be to combine some of these parameters in a "master" classifier. Also the training set could be explanded by using a larger fraction of the Hubble Legacy Archive. Now it only runs and evaluates on the five CANDELS fields.

```python
import os
import glob
import numpy as np
from numpy import sqrt, fromfunction

import matplotlib
import matplotlib.pylab as plt
import pandas as pd
pd.set_option('max_columns', None)
import requests
from io import BytesIO
import astropy
from astropy.io import fits
from astropy.stats import sigma_clipped_stats
from astropy import units as u
from astropy.coordinates import SkyCoord

import PIL
from PIL import Image
from photutils import CircularAperture
from photutils import aperture_photometry
from photutils import DAOStarFinder
from pyvo.dal import tap

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing import image
import sklearn

HST_TAP_CAT = "http://vao.stsci.edu/HSCTAP/tapservice.aspx"
tapcathst = tap.TAPService(HST_TAP_CAT)
GAIAMOCK_TAP_CAT = "https://dc.zah.uni-heidelberg.de/__system__/tap/run/tap"
tapcatmock = tap.TAPService(GAIAMOCK_TAP_CAT)
ESA_TAP_CAT = "https://gaia.aip.de/tap"
tapcatesa = tap.TAPService(ESA_TAP_CAT)

def query_gaia3(ra_search, dec_search, radius):
    query="""SELECT source_id, ra, dec, parallax, pmra, pmdec, 
    phot_g_mean_mag, phot_bp_mean_mag, phot_rp_mean_mag, 
    astrometric_excess_noise, astrometric_excess_noise_sig, ruwe, astrometric_sigma5d_max
    FROM gaiaedr3.gaia_source WHERE CONTAINS(point('ICRS', ra, dec), circle('ICRS', %f, %f, %f))=1""" % (ra_search, dec_search, radius)
    res = tapcatesa.run_sync(query)
    df_Gaia = res.to_table().to_pandas()
    return df_Gaia

def query_gaia3_heidelberg(ra_search, dec_search, radius):
    query="""SELECT source_id, ra, dec, parallax, pmra, pmdec, 
    phot_g_mean_mag, phot_bp_mean_mag, phot_rp_mean_mag, 
    astrometric_excess_noise, ruwe, t.fidelity_v1
    FROM gaia.edr3lite JOIN gedr3spur.main as t using (source_id) WHERE CONTAINS(point('ICRS', ra, dec), circle('ICRS', %f, %f, %f))=1""" % (ra_search, dec_search, radius)
    res = tapcatmock.run_sync(query)
    df_Gaia = res.to_table().to_pandas()
    return df_Gaia

def query_hla_candels():
    """Query Hubble Legacy Archive database for catalogue data, CANDELS fields"""  
    query="""SELECT * FROM dbo.DetailedCatalog
    WHERE ((PropID ='12063') OR (PropID ='12440') OR (PropID ='12443') OR (PropID ='12061') OR (PropID ='12064'))
    AND Filter = 'F160W'
    AND MagAuto < 22."""
    print(query)
    res = tapcathst.run_sync(query)
    df_HLA = res.to_table().to_pandas()
    return df_HLA

def query_hla(nobjects):
    """Query Hubble Legacy Archive database for catalogue data"""
    query=f"""SELECT TOP {nobjects} * FROM dbo.DetailedCatalog
    WHERE Filter = 'F160W'
    AND MagAuto < 22."""
    print(query)
    res = tapcathst.run_sync(query)
    df_HLA = res.to_table().to_pandas()
    return df_HLA

def getimage(x,y,image):
    """Retrieve corresponding image, based on x, y coordinates and image name"""
    thisurl = "https://hla.stsci.edu/cgi-bin/fitscut.cgi?red="+image+"&amp;size=99&amp;x="+str(x)+"&amp;y="+str(y)+"&amp;format=fits&amp;Badpix=True"
    r = requests.get(thisurl)
    hdu = fits.open(BytesIO(r.content))
    return hdu

def get_star(im):
    mask = np.isnan(im)
    daofind = DAOStarFinder(fwhm=3.0, threshold=1.,roundlo=-2., roundhi=2.)  
    sources = daofind(im,mask).to_pandas()
    sources['distance'] = np.sqrt((sources['xcentroid']-48.)**2+(sources['ycentroid']-48.)**2)
    SELCENTER = sources['distance'] < 14.
    SOURCESCENTER = sources[SELCENTER]
    SOURCESCENTER = SOURCESCENTER.sort_values(['peak'],ascending=False).reset_index()
    #print(sources)
    xpos = SOURCESCENTER['xcentroid'][0]
    ypos = SOURCESCENTER['ycentroid'][0]
    return xpos, ypos, SOURCESCENTER

def curve_of_growth(im, radii, xpos, ypos):
    #xpos, ypos = get_star(im)
    aperture = [CircularAperture((xpos,ypos), r=r) for r in radii]
    phot_table = aperture_photometry(im, aperture).to_pandas()
    return phot_table.iloc[0, 3:103]/phot_table.iloc[0, 102]

def get_background(im):
    """Measure rough background in image from the edges"""
    edges=[im[0,:], im[:,-1], im[-1,:], im[:,0]]
    background = sigma_clipped_stats(edges, sigma=3, maxiters=10)[0]
    return background

def dist_circle(n, xcen=None ,ycen=None):
    """
    Created on Dec 10, 2012
    Form a square array where each value is its distance to a given center.
    @author: jweingrill@aip.de

    INPUTS:
      N = either  a scalar specifying the size of the N x N square output
               array, or a 2 element vector specifying the size of the
               N x M rectangular output array.

    OPTIONAL INPUTS:
      XCEN,YCEN = Scalars designating the X,Y pixel center.  These need
               not be integers, and need not be located within the
               output image.   If not supplied then the center of the output
               image is used (XCEN = YCEN = (N-1)/2.).

    OUTPUTS:
       IM  - N by N (or M x N) floating array in which the value of each 
               pixel is equal to its distance to XCEN,YCEN
    """
    if type(n) == int:
        ny = n
        nx = n
    elif len(n) == 2:
        nx = n[0]
        ny = n[1]
    else:
        exit('ERROR - Output size parameter N must contain 1 or 2 elements')
        
    if xcen is None:
        xcen = (nx-1.)/2.
    if ycen is None:
        ycen = (ny-1.)/2.
        
    im = fromfunction(lambda x,y: sqrt((x-xcen)**2+(y-ycen)**2), (nx,ny))
    return(im)
```

# Prepare images to execute CNN model

```python
# This can be generalised to data sets outside of CANDELS fields
files = glob.glob('./CNN_input/Run/*')
for f in files:
    os.remove(f)
        
source = './cutouts_HLA_CANDELS_orig/'
files = os.listdir(source)
hnewdim = 16 #half the dimension. The image files are 32x32 pixels
for i, file in enumerate(files):
    hdul = fits.open(source+file)
    orig = hdul[0].data
    holddim = 0.5*len(orig)
    cutout = orig[int(holddim-hnewdim):int(holddim+hnewdim),int(holddim-hnewdim):int(holddim+hnewdim)]
    nnan = np.isnan(cutout).sum()
    if nnan < 0.5:
        cutout = cutout/np.max(cutout)*255. #scale from 0 to 255
        im = Image.fromarray(cutout)
        if im.mode != 'RGB':
            im = im.convert('RGB')
        im.save('./CNN_input/Run/'+file.split('.')[0]+'.jpeg')
```

# Run model and make predictions

```python
# Load model that was trained in previous notebook
model = keras.models.load_model('cnn_model_binaryclassifier')
```

```python
def predictImage(filename):
    """Can be different image than is part of test and train sets.
    This is not optimised for speed, since it runs on a single image.
    Input has to be jpeg image, no FITS."""
    if not os.path.isfile(filename):
        return(np.zeros((1,1))-1)
    img1 = image.load_img(filename,target_size=(32,32))
    plt.imshow(img1)
    Y = image.img_to_array(img1)
    X = np.expand_dims(Y,axis=0)
    XNORM = X/255. #From 0 to 1
    val = model.predict(XNORM)
    #print(val)
    if val == 1:     
        plt.xlabel("Single",fontsize=30)
    elif val == 0:
        plt.xlabel("Binary",fontsize=30)
    return(val)
```

```python
# Fill Dataframe with predictions from CNN
df_new['binary_classifier'] = -1 # -1 when no information available (when there are NaNs in the cutout)
for i in range(len(df_new)):
    label_predict = predictImage('./CNN_input/Run/df_new_entry_'+str(i)+'.jpeg')
    df_new.loc[i,'binary_classifier'] = label_predict[0][0]
```

```python
# Extended source flag based on circular aperture fluxes
# determine which are usable and have no NaN near the centre, and determine centroids
DISTCIRC = dist_circle((99,99), 48., 48.)
SELCENTRE = (DISTCIRC < 15.)

df_new['hststamp_xpix'] = float("NaN")
df_new['hststamp_ypix'] = float("NaN")
df_new['hststamp_sharpness'] = float("NaN")
df_new['hststamp_roundness1'] = float("NaN")
df_new['hststamp_roundness2'] = float("NaN")

nnotgood = 0.
for i in range(len(df_new)):
    hdu = fits.open('./cutouts_HLA_CANDELS_orig/df_new_entry_'+str(i)+'.fits') #background is removed
    im = hdu[0].data
    
    if (np.isnan(im[SELCENTRE]).sum() < 0.5): #this also throws out saturated stars
        xpos, ypos, sources = get_star(im)
        df_new.loc[i, ('hststamp_xpix')] = xpos
        df_new.loc[i, ('hststamp_ypix')] = ypos
        df_new.loc[i, ('hststamp_sharpness')] = sources['sharpness'][0]
        df_new.loc[i, ('hststamp_roundness1')] = sources['roundness1'][0]
        df_new.loc[i, ('hststamp_roundness2')] = sources['roundness2'][0]

    else:
        nnotgood+=1
        
print(f'{nnotgood:.0f} sources have NaNs in the image cutouts')
```

```python
# get curve of growth for each usable image (no NaN)
df_new['aperture_5_10'] = float("NaN")
df_new['aperture_10_20'] = float("NaN")
df_new['aperture_20_60'] = float("NaN")
df_new['HST_good_star'] = float("NaN")

radii = np.linspace(0.1,14,100)
ALLCOGS = np.ndarray((len(df_new),len(radii)))
ALLCOGS[:] = float("NaN")
for i in (1-(np.isnan(df_new.loc[:,'hststamp_xpix']))).to_numpy().nonzero()[0]: #loop over indices without NaN
    hdu = fits.open('./cutouts_HLA_CANDELS_orig/df_new_entry_'+str(i)+'.fits') #background is removed
    im = hdu[0].data
    xpos = df_new.loc[i,'hststamp_xpix']
    ypos = df_new.loc[i,'hststamp_ypix']
    cog = curve_of_growth(im, radii, xpos, ypos)
    #print(cog)
    ALLCOGS[i,:] = cog
    df_new.loc[i,'aperture_5_10']  = cog[10]/cog[5]
    df_new.loc[i,'aperture_10_20'] = cog[20]/cog[10]
    df_new.loc[i,'aperture_20_60'] = cog[60]/cog[20]
    df_new.loc[i,'HST_good_star'] = 0 

# This is the selection:    
SEL = (df_new['aperture_5_10'] < 3.05) & (df_new['aperture_10_20'] < 2.20) & (df_new['aperture_20_60'] < 1.45)
df_new.loc[SEL,'HST_good_star'] = 1
```

```python
# Check how well the two different classifiers work.
# Both figures show (aperture magnitude - total magnitude) versus total magnitude in the F160W filter.
# Colour scale shows the value of the classifier.
# The first classifier works well in identifying most extended sources (such as galaxies)
# The second classifier misses many of these extended sources, but is tuned to identifying closely separated binaries.

X = df_new.MagAuto
Y = df_new.MagAper2-df_new.MagAuto
plt.scatter(X, Y, marker='.',c=df_new['HST_good_star'], alpha=0.5)
plt.ylabel('Apermag-Totalmag')
plt.xlabel('Totalmag')
plt.title('Curve-of-growth classifier')
cbar = plt.colorbar()
cbar.set_label('Classifier') 
plt.ylim(0,0.6)
plt.show()

X = df_new.MagAuto
Y = df_new.MagAper2-df_new.MagAuto
SEL = df_new['binary_classifier'] >=0
plt.scatter(X[SEL], Y[SEL], marker='.',c=df_new.loc[SEL,'binary_classifier'], alpha=0.5)
plt.ylabel('Apermag-Totalmag')
plt.xlabel('Totalmag')
plt.title('Binary classifier (CNN)')
cbar = plt.colorbar()
cbar.set_label('Classifier') 
plt.ylim(0,0.6)
plt.show()
```

# Different Gaia parameters 
Short description of the different Gaia parameters that are considered here, taken from Gaia documentation:
#### astrometric_excess_noise
This is the excess noise of the source, measuring the disagreement, expressed as an angle, between the observations of a source and the best-fitting standard astrometric model (using five astrometric parameters). A value of 0 signifies a well-behaved source, a positive value signifies that the residuals are larger than expected. 
#### astrometric_excess_noise_sig
Significance of excess noise (double) A dimensionless measure (D) of the significance of the calculated astrometric_excess_noise. A value D>2 indicates that the given excess noise is probably significant. For good fits in the limit of a large number of observations, D should be zero in half of the cases and approximately follow the positive half of a normal distribution with zero mean and unit standard deviation for the other half. Consequently, D is expected to be greater than 2 for only a few percent of the sources with well-behaved astrometric solutions.
#### astrometric_sigma5d_max
The longest principal axis in the 5-dimensional error ellipsoid. This is a 5-dimensional equivalent to the semi-major axis of the position error ellipse and is therefore useful for filtering out cases where one of the five parameters, or some linear combination of several parameters, is particularly ill-determined. 
#### RUWE (see Belokurov+20, Unresolved stellar companions with Gaia DR2 astrometry )
Renormalised Unit Weight Error (RUWE) associated to each source in gaia_source.
The RUWE is expected to be around 1.0 for sources where the single-star model provides a good fit to the astrometric observations. A value significantly greater than 1.0 (say, >1.4) could indicate that the source is non-single or otherwise problematic for the astrometric solution. 
#### fidelity_v1
A classifier for spurious astrometric solutions in Gaia eDR3, described in Rybizki+22: "A classifier for spurious astrometric solutions in Gaia eDR3"

```python
# Set this parameter:
allowed_contamination_fraction = 0.04
```

```python
# Here we only use the curve-of-growth result, not the binary classifier
HSTSTAR = (df_new['HST_good_star'] > 0.5)
HSTNOSTAR = (df_new['HST_good_star'] < 0.5)
purity_tot = (HSTSTAR.sum()/(HSTSTAR.sum()+HSTNOSTAR.sum()))
plt.plot([0,1],[1.-purity_tot,1.-purity_tot], '--', label='purity total cat')
print(f'Fraction of HST-identified crap = {1.0-purity_tot:.3f}')
#completeness is defined as keeping all stars that look good. So the ideal spot on the diagram is at position (1,0)

COLOURS = ['black','red','blue','green','purple','cyan']
PARAMETERS = ['astrometric_excess_noise','astrometric_excess_noise_sig','ruwe','fidelity_v1','astrometric_sigma5d_max'] #,'d2dself']
for parameter, color in zip(PARAMETERS,COLOURS):
    VALUES = df_new[parameter].dropna().sort_values().unique()
    POSX = []
    POSY = []
    xbest = -1.
    ybest = -1.
    parameterbest = -1.
    for value in df_new[parameter].dropna().sort_values().unique():
        if (parameter == 'fidelity_v1') or (parameter == 'd2dself'): # In that case high values indicate good sources
            SELHSTSTAR = (df_new['HST_good_star'] > 0.5) & (df_new[parameter] > value)
            SELHSTNOSTAR = (df_new['HST_good_star'] < 0.5) & (df_new[parameter] > value)
        else: # In that case high values indicate crap sources
            SELHSTSTAR = (df_new['HST_good_star'] > 0.5) & (df_new[parameter] < value)
            SELHSTNOSTAR = (df_new['HST_good_star'] < 0.5) & (df_new[parameter] < value)
        if SELHSTSTAR.sum()+SELHSTNOSTAR.sum() >= 0.5: # make sure don't divide by zero
            xnew = SELHSTSTAR.sum()/HSTSTAR.sum()
            ynew = 1.-SELHSTSTAR.sum()/(SELHSTSTAR.sum()+SELHSTNOSTAR.sum())
            POSX.append(xnew)
            POSY.append(ynew)
            if ynew < allowed_contamination_fraction:
                if (parameter == 'fidelity_v1') or (parameter == 'd2dself'):
                    if xbest < -0.5: # then it hasn't been set yet, keep first value
                        print('yeadh')
                        xbest = xnew
                        ybest = ynew 
                        parameterbest = value
                else: # keep last value
                    xbest = xnew
                    ybest = ynew 
                    parameterbest = value
    print(f'can cut {parameter} at {parameterbest:.5f} to get contamination of {ybest:.3f} and completeness of {xbest:.3f}')
    plt.plot(POSX, POSY, color=color, label=parameter, alpha=0.4)
    
plt.xlabel('Completeness')
plt.ylabel('Contamination')
plt.legend()
plt.savefig('compare_Gaia_parameters.png',dpi=200)
plt.show()
```
